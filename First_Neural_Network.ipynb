```python
import torch
from torch import nn
import random
```

```python
def mystery(a,b):
  return torch.tensor(a+3*b)
```

```python
model = nn.Sequential(nn.Linear(2,1)) # nn w/ one neuron, 2 inputs, 1 output
model
```

```python
# train nn
criterion = nn.MSELoss() # how well nn is doing, since using values use MSE which gives value for how much nn is producing values that are different than what it should be
optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9) # how nn is updated during training process
```

```python
for i in range(1000):

  a = random.random()
  b = random.random()
  desiredOutput = mystery(a,b)

  # to train network: pass two inputs into our network
  output = model(torch.tensor([a,b]))

  # compare how this output is compared to desired output
  loss = criterion(output.squeeze(), desiredOutput)

  # track loss
  # print out loss every 100 times
  if (i % 100)==0:
    print(f"Loss: {loss.item()}")

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  # can see loss decreases so nn has learned how to map the inputs to the outputs
```

```python
# test nn
a = 1.0
b = -1.0
output = model(torch.tensor([a,b]))
print(output.item())
```

```python
mystery(a,b)
# without any knowledge about mystery function, nn was able to determine what the mystery function was by looking at input and output pairs
```
